```python
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('seaborn-bright')
```

```python
densidadeJ3 = pd.read_csv('DensidadesJ3.dat', sep='\t')
densidadeJ4 = pd.read_csv('DensidadesJ4.dat', sep='\t')

densidadeT3 = pd.read_csv('DensidadesT3.dat', sep='\t')
densidadeT4 = pd.read_csv('DensidadesT4.dat', sep='\t')
```

```python
fig, ax = plt.subplots(1,1, figsize = (10,7))
ax.scatter(densidadeJ3['Temperatura (K)'], densidadeJ3['Densidade (g/cm)'], label = 'James tip3p', s = 50, edgecolor = 'blue', facecolor = 'none')
ax.scatter(densidadeT3['Temperatura (K)'], densidadeT3['Densidade (g/cm)'], label = 'Turma tip3p', marker = '>', edgecolor = 'blue', facecolor = 'none')

ax.scatter(densidadeJ4['Temperatura (K)'], densidadeJ4['Densidade (g/cm)'], label = 'James tip4p', s = 50 ,  edgecolor = 'lime', facecolor = 'none')
ax.scatter(densidadeT4['Temperatura (K)'], densidadeT4['Densidade (g/cm)'], label = 'Turma tip4p', marker = '>',  edgecolor = 'lime', facecolor = 'none')

ax.plot(densidadeT4['Temperatura (K)'], densidadeT4['Densidade (g/cm)'], color = 'lime')
ax.plot(densidadeT3['Temperatura (K)'], densidadeT3['Densidade (g/cm)'], color = 'blue')

#ax.axvline(200, linestyle = '--', c = 'blue', label = 'Transio de fase tip3p')
#ax.axvline(300, linestyle = '--', c = 'lime', label = 'Transio de fase tip4p')

ax.set_ylabel('Densidade (g/cm)', fontsize = 16)
ax.set_xlabel('Temperatura (K)', fontsize = 16)
ax.set_title('Densidades', fontsize = 24)
ax.grid('--')
ax.legend(fontsize = 13)
plt.savefig('Densidades.png', dpi = 200)
```

```python
energiaJ3 = pd.read_csv('EnergiasJ3.dat', sep='\t')
energiaJ4 = pd.read_csv('energiasJ4.dat', sep='\t')

energiaT3 = pd.read_csv('EnergiasT3.dat', sep='\t')
energiaT4 = pd.read_csv('EnergiasT4.dat', sep='\t')
```

```python
fig, ax = plt.subplots(1,1, figsize = (10,7))
ax.scatter(energiaJ3['Temperatura (K)'], energiaJ3['Etot (kcal/mol)'], label = 'James tip3p', s = 50, edgecolor = 'blue', facecolor = 'none')
ax.scatter(energiaT3['Temperatura (K)'], energiaT3['Etot (kcal/mol)'], label = 'Turma tip3p', marker = '>', edgecolor = 'blue', facecolor = 'none')

ax.scatter(energiaJ4['Temperatura (K)'], energiaJ4['Etot (kcal/mol)'], label = 'James tip4p', s = 50 ,  edgecolor = 'lime', facecolor = 'none')
ax.scatter(energiaT4['Temperatura (K)'], energiaT4['Etot (kcal/mol)'], label = 'Turma tip4p', marker = '>',  edgecolor = 'lime', facecolor = 'none')

ax.plot(energiaT4['Temperatura (K)'], energiaT4['Etot (kcal/mol)'], color = 'lime')
ax.plot(energiaT3['Temperatura (K)'], energiaT3['Etot (kcal/mol)'], color = 'blue')

ax.set_ylabel('Enegia total (kcal/mol)', fontsize = 16)
ax.set_xlabel('Temperatura (K)', fontsize = 16)
ax.set_title('Energia', fontsize = 24)
ax.grid('--')
ax.legend(fontsize = 13)
plt.savefig('energias.png', dpi = 200)
```
