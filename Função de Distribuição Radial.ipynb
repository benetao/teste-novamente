```

# List of file paths
temperaturas = [180 + 10*i for i in range(5)]
file_paths = [f"{i}/rdf-cpptraj-WAT-O-O.dat" for i in temperaturas]

for i in range(len( file_paths)):
    file_paths[i]= "pdf_tip3p/" + file_paths[i]

cval = np.linspace(0,1,len(file_paths))
cores = plt.get_cmap("plasma")(cval)


fig, ax = plt.subplots(1,1)
k = 0
for file_path in file_paths:
    x, y = [], []
    with open(file_path, 'r') as file:
        # Skip the two header lines
        next(file)
        next(file)
        for line in file:
            values = line.split()
            x.append(float(values[0]))
            y.append(float(values[1]))
    label=file_path.replace('/rdf-cpptraj-WAT-O-O.dat',' K')
    label=label.replace('pdf_tip3p/','')
    y = np.array(y)
    y = y-k*3

    ax.plot(x, y, label=label, color = cores[k])
    ax.set_xlabel('Distância (Å)')
    ax.set_ylabel(r'G(r) (Å$^{-2}$)')
    ax.set_title('Transição de fase tip3p')

    k+=1

#plt.title('Plot of Data from Multiple Files')
ax.legend(loc='lower right',  ncol=1, fancybox=True, shadow=True)
ax.set_yticks([])  

ax.grid('--')

plt.savefig('PDFs_tip3p.png', dpi = 200)
```

```
import numpy as np

def conjugate_gradient(A, b, tol=1e-6, max_iter=None):
    n = len(b)
    x = np.zeros(n)  # Inicialize x com um vetor de zeros
    r = b - np.dot(A, x)  # Resíduo inicial
    p = r  # Direção de pesquisa inicial
    rsold = np.dot(r, r)

    if max_iter is None:
        max_iter = n  # Define um número máximo de iterações como o tamanho do sistema

    for i in range(max_iter):
        Ap = np.dot(A, p)
        alpha = rsold / np.dot(p, Ap)
        x = x + alpha * p
        r = r - alpha * Ap
        rsnew = np.dot(r, r)
        if np.sqrt(rsnew) < tol:
            break
        p = r + (rsnew / rsold) * p
        rsold = rsnew

    return x
```

```
A = np.array([[3, 1, -1], [1, 4, 2], [-1, 2, 4]])
b = np.array([1, -1, 1])
x = conjugate_gradient(A, b)
print("Solução x:", x)
```
